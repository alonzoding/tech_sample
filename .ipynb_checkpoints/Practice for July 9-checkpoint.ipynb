```python
nypd["BORO"].value_counts().plot(kind = "bar")
plt.title("shooting incidents per borough")
plt.xlabel("boroughs")
plt.ylabel("number of shootings")
```

```python
boro_filter = nypd["BORO"] == "BRONX"
race_filter = nypd["VIC_RACE"] == "WHITE HISPANIC"
nypd[boro_filter & race_filter]
```

```python
nypd[(nypd["OCCUR_TIME"] >= "09:00:00") & (nypd["OCCUR_TIME"] <= "21:00:00")].sort_values("OCCUR_TIME")
```

```python
nypd[(nypd["OCCUR_TIME"] < "09:00:00") | (nypd["OCCUR_TIME"] > "21:00:00")].sort_values("OCCUR_TIME")
```

```python
day = nypd[(nypd["OCCUR_TIME"] >= "09:00:00") & (nypd["OCCUR_TIME"] <= "21:00:00")].sort_values("OCCUR_TIME")
day["OCCUR_TIME"].value_counts().sort_index().plot()
```

```python
night = nypd[(nypd["OCCUR_TIME"] < "09:00:00") | (nypd["OCCUR_TIME"] > "21:00:00")].sort_values("OCCUR_TIME")
night["OCCUR_TIME"].value_counts().sort_index().plot()
```
